# Celda 1: Instalaci√≥n y Configuraci√≥n
Instalamos ultralytics (la librer√≠a oficial de YOLO) y conectamos el Drive.

```python
# 1. INSTALACI√ìN DE LIBRER√çAS
# Instalamos la librer√≠a oficial de YOLOv8
print("--- Instalando Ultralytics (YOLOv8) ---")
%pip install ultralytics -q

import ultralytics
from ultralytics import YOLO
import os
import shutil
from google.colab import drive

# Verificamos que la instalaci√≥n fue correcta y que detecta la GPU
ultralytics.checks()

# Montamos Drive para recuperar los datos y guardar los modelos al final
print("\n--- Conectando con Google Drive ---")
drive.mount('/content/drive')

# Rutas
DRIVE_FOLDER = '/content/drive/MyDrive/Practica_ISIC'
ZIP_NAME = 'dataset_ISIC.zip'
```

# Celda 2: Carga de Datos
Traemos los datos del ZIP y los descomprimimos.

```python
# 2. PREPARACI√ìN DEL DATASET
print("--- Cargando datos desde Drive ---")

# Ruta origen (Drive) y destino (Colab)
zip_path_drive = os.path.join(DRIVE_FOLDER, ZIP_NAME)

# Copiamos el zip al entorno local de Colab
if not os.path.exists(ZIP_NAME):
    print(f"Copiando {ZIP_NAME}...")
    shutil.copy(zip_path_drive, '/content/')

    print("Descomprimiendo dataset...")
    !unzip -q {ZIP_NAME}
    print("¬°Datos listos en la carpeta 'dataset'!")
else:
    print("El archivo ZIP ya estaba en el entorno. Se omite la copia.")

# Verificaci√≥n r√°pida
if os.path.exists('dataset/train') and os.path.exists('dataset/test'):
    print(f"Verificaci√≥n: OK. Carpetas train/test encontradas.")
else:
    print("ERROR: No se encuentra la estructura de carpetas esperada.")
```

# Celda 3: Definici√≥n de Hiperpar√°metros
Configuramos las reglas del entrenamiento seg√∫n la gu√≠a.


**Epochs: 20** (Seg√∫n hemos visto, entre 20 y 50, suele ser suficiente para esta cantidad de im√°genes. Nosotros hemos utilizado 20, aunque para ejecutar train.ipynb en el v√≠deo lo hemos puesto a 1. Igualmente eval.ipynb en el v√≠deo tomar√° el zip generado con 'epochs': 20, por eso el microcorte que habr√° en el v√≠deo).


**Imgz: 224** (Tama√±o est√°ndar para clasificaci√≥n ).


**Batch: 16** (Im√°genes procesadas a la vez ).

```python
# 3. CONFIGURACI√ìN
# Definimos los par√°metros comunes para todos los entrenamientos

ARGS = {

    'data': 'dataset',      # Carpeta donde est√°n las im√°genes

    'epochs': 1,           # N√∫mero de pasadas completas

    'imgsz': 224,           # Tama√±o de la imagen (est√°ndar clasificaci√≥n)

    'batch': 16,            # Tama√±o del lote

    'project': 'ISIC_Project', # Carpeta principal de resultados

    'exist_ok': True        # Sobrescribir si existe
}

print("Configuraci√≥n de entrenamiento definida.")
```

# Celda 4: Entrenamiento Modelo NANO (n)
El modelo m√°s ligero.

```python
# 4. ENTRENAMIENTO: YOLOv8 NANO
print("--- Iniciando entrenamiento: YOLOv8n-cls (Nano) ---")

# Cargamos el modelo pre-entrenado
model_n = YOLO('yolov8n-cls.pt')

# Entrenamos
# name='yolov8n_run' crea una subcarpeta espec√≠fica para este modelo
results_n = model_n.train(**ARGS, name='yolov8n_run')

print("\n‚úÖ Entrenamiento Nano finalizado.")
```

#Celda 5: Entrenamiento Modelo SMALL (s)
Un poco m√°s pesado, te√≥ricamente m√°s preciso.

```python
# 5. ENTRENAMIENTO: YOLOv8 SMALL
print("--- Iniciando entrenamiento: YOLOv8s-cls (Small) ---")

# Cargamos el modelo pre-entrenado 's'
model_s = YOLO('yolov8s-cls.pt')

# Entrenamos
results_s = model_s.train(**ARGS, name='yolov8s_run')

print("\n‚úÖ Entrenamiento Small finalizado.")
```

#Celda 6: Entrenamiento Modelo MEDIUM (m)
El m√°s grande. Seg√∫n hemos visto este podr√≠a sufrir sobreajuste (memorizar en vez de aprender) porque tenemos pocas im√°genes. Hay que observar si la p√©rdida de validaci√≥n (val_loss) empieza a subir al final.

```python
# 6. ENTRENAMIENTO: YOLOv8 MEDIUM
print("--- Iniciando entrenamiento: YOLOv8m-cls (Medium) ---")

# Cargamos el modelo pre-entrenado 'm'
model_m = YOLO('yolov8m-cls.pt')

# Entrenamos
results_m = model_m.train(**ARGS, name='yolov8m_run')

print("\n‚úÖ Entrenamiento Medium finalizado.")
```

#Celda 7: Guardar Resultados
Empaquetamos todos los modelos entrenados y gr√°ficas que YOLO ha generado en la carpeta ISIC_Project.

```python
# 7. GUARDAR RESULTADOS EN DRIVE
print("--- Empaquetando modelos y gr√°ficas ---")

output_zip = 'modelos_entrenados.zip'
project_folder = 'ISIC_Project' # La carpeta que definimos en ARGS

# Comprimimos la carpeta del proyecto que contiene las 3 ejecuciones (n, s, m)
!zip -r -q {output_zip} {project_folder}

# Movemos a Drive
dest_path = os.path.join(DRIVE_FOLDER, output_zip)
print(f"Guardando {output_zip} en Drive...")
shutil.copy(output_zip, dest_path)

print(f"\n‚úÖ ¬°PROCESO TERMINADO! Modelos guardados en: {dest_path}")
```
